## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
from typing import Literal

def convert_format(source: str, format: Literal['snake', 'pascal'] = None) -> str:
    if source.count('_') > 0 or format == 'snake':
        return ''.join(map(lambda x: x.capitalize(), source.split('_')))
    
    words = []
    buffer = ''
    for idx, char in enumerate(source):
        if len(buffer) == 0 or char.islower():
            buffer += char
        else:
            words.append(buffer)
            buffer = char
        if idx == len(source) - 1:
            words.append(buffer)
    return '_'.join(words)
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def validate_date(value: str) -> bool:
    parts = value.split('.')
    if len(parts) != 3:
        return False
    
    day, month, year = parts

    if not year.isdecimal() or int(year) < 1:
        return False

    if not month.isdecimal() or int(month) < 1 or int(month) > 12:
        return False
    
    max_day = 31 if int(month) in [1, 3, 5, 7, 8, 10, 12] else 30
    if int(month) == 2:
        is_leap_year = int(year) % 4 == 0 and (int(year) % 100 != 0 or int(year) % 400 == 0)
        max_day = 29 if is_leap_year else 28

    return day.isdecimal() and int(day) >= 1 and int(day) <= max_day

    
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def validate_simple_number(value: int) -> bool:
    if value < 2:
        return False
    for i in range(2, int(value**0.5) + 1):
        if value % i == 0:
            return False
    return True
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def is_valid_name(name: str) -> bool:
    return len(name) > 0 and name.isalpha()

def is_valid_age(age: str) -> bool:
    return age.isdecimal() and int(age) >= 18 and int(age) <= 60

def is_valid_ID(ID: str) -> bool:
    return ID.isdecimal() and int(ID) > 0

def format_ID(ID: str) -> str:
    return ID.zfill(8)

def print_users(users: dict) -> None:
    if not users:
        print("Нет данных о пользователях.")
        return
    
    headers = ['ID', 'Имя', 'Фамилия', 'Возраст']
    print('\n' + '-' * 50)
    print(f"{headers[0]:<10} {headers[1]:<15} {headers[2]:<15} {headers[3]:<6}")
    print("-" * 50)
    
    for user_id in sorted(users.keys()):
        first_name, last_name, age = users[user_id]
        print(f'{user_id:<10} {first_name:<15} {last_name:<15} {age:<6}')
    print('-' * 50)

users = {}

while True:
    user_input = input("Введите данные о пользователе: ").strip()
    if not user_input:
        break

    parts = user_input.split()
    if len(parts) != 4:
        print("Ошибка: нужно ввести ровно 4 значения (имя, фамилия, возраст, ID).")
        continue

    first_name_raw, last_name_raw, age_raw, ID_raw = parts

    if not is_valid_name(first_name_raw):
        print(f"Ошибка: имя '{first_name_raw}' должно состоять только из букв.")
        continue

    if not is_valid_name(last_name_raw):
        print(f"Ошибка: фамилия '{last_name_raw}' должна состоять только из букв.")
        continue

    if not is_valid_age(age_raw):
        print(f"Ошибка: возраст '{age_raw}' должен быть числом от 18 до 60.")
        continue

    if not is_valid_ID(ID_raw):
        print(f"Ошибка: ID '{ID_raw}' должен быть положительным целым числом.")
        continue

    first_name = first_name_raw.capitalize()
    last_name = last_name_raw.capitalize()
    age = int(age_raw)
    ID = format_ID(ID_raw)

    if ID in users:
        print(f"Ошибка: ID '{ID}' уже существует. Введите уникальный ID.")
        continue

    users[ID] = (first_name, last_name, age)

print_users(users)

```
