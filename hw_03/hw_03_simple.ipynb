## Задание 1

Написать функцию, которая принимает неограниченное количество чисел в виде позиционных аргументов и ключевой аргумент — операцию над этими числами (сложение или умножение).
Функция должна возвращать результат выполнения указанной операции.

````python
from typing import Literal

def calc(*args: float, operation: Literal['*', '+']) -> float:
    result = None
    for element in args:
        if result is None:
            result = element
            continue

        if operation == '+':
            result += element
        elif operation == '*':
            result *= element
            
    return result
````

## Задание 2

Написать функцию для ввода из консоли целого числа.
Если введено не число, функция должна вывести соответствующее сообщение и предложить ввести значение заново — до тех пор, пока пользователь не введёт корректное число.


````python
def input_num() -> int:
    while True:
        user_input = input('Введите целое число: ')
        
        minus_count = user_input.count('-')
        plus_count = user_input.count('+')

        if (minus_count <= 1 and plus_count <= 1 and
            (minus_count == 0 or user_input[0] == '-') and
            (plus_count == 0 or user_input[0] == '+') and
            user_input.replace('-', '').replace('+', '').isdecimal()):
            return int(user_input)
        print('Число введено неправильно')
````

## Задание 3

Написать функцию, которая создаёт абсолютный путь к файлу.

Позиционные аргументы:
- название диска,
- неограниченное количество папок,
- имя файла (без расширения).

Ключевые аргументы:
- ext — расширение файла,
- sep — разделитель (по умолчанию '/').

Пример:
```full_path('c:', 'work', 'python', 'function', 'main', ext='py') ➜ 'c:/work/python/function/main.py'```


````python
def full_path(*args: str, ext: str, sep: str = '/'):
    return sep.join(args) + f'.{ext}'
````

## Задание 4

Написать функцию, которая принимает список, состоящий из объектов разных типов, и возвращает словарь, где:
- ключи — типы данных объектов;
- значения — списки объектов соответствующего типа.


````python
def get_dict(ls: list[any]) -> dict[str, list[any]]:
    result = {}
    for element in ls:
        result.setdefault(type(element), []).append(element)
    return result

````
