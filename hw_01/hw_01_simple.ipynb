## Задание 1

Пользователь в бесконечном цикле вводит числа через консоль до тех пор, пока не введёт что-то, отличное от числа (после этого цикл завершается). В конце программа должна вывести сумму всех введённых чисел.

```python
result = 0
while True:
    user_input = input('Введите число: ').strip()

    minus_count = user_input.count('-')
    plus_count = user_input.count('+')
    dot_count = user_input.count('.')

    if (minus_count > 1 or plus_count > 1 or dot_count > 1 or
        (minus_count == 1 and user_input[0] != '-') or
        (plus_count == 1 and user_input[0] != '+') or
        not user_input.replace('-', '').replace('+', '').replace('.', '').isdecimal()):
        break

    result += float(user_input)
print(result)
```

## Задание 2

Пользователь вводит строку через консоль (состоящую только из букв в разном регистре). Программа должна создать новую строку, содержащую только те буквы из исходной строки, которые были в верхнем регистре.



```python
user_input = input('Введите строку из букв: ')

result = ''
for char in user_input:
    if char.isupper():
        result += char
print(f'Исходная строка: {user_input}\nРезультат: {result}')
```

## Задание 3

Классическая игра FizzBuzz. Программа в цикле выводит в консоль числа от 1 до 100.

- Если число кратно 3, вместо него выводится Fizz.
- Если кратно 5 — Buzz.
- Если кратно и 3, и 5 — FizzBuzz.



```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

## Задание 4

Пользователь вводит строку, содержащую любые символы (буквы, цифры и специальные символы). Программа должна вычислить и вывести сумму всех цифр, присутствующих в этой строке.



```python
user_input = input('Введите строку: ')

result = 0
for char in user_input:
    if char.isdecimal():
        result += int(char)
print(f'Сумма цифр в строке "{user_input}": {result}')
```

## Задание 5

Пользователь вводит через консоль 10 чисел. Программа должна определить максимальное число и вывести его в консоль.


```python
max = None
for i in range(1, 11):
    num = float(input('Введите число: '))
    print(f'Число {i}: {num}')
    if max is None or num > max:
        max = num
print(f'Максимальное число: {max}')
    
```
