# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr_1d = np.zeros(10)
arr_1d[4] = 1
print("Одномерный массив: ", arr_1d, sep='\n')
arr_2d = arr_1d.reshape(5, -1)
print("Двумерный массив: ", arr_2d, sep='\n')
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.arange(10, 50)
reverse_arr = arr[::-1]
print(reverse_arr)
print("Четные элементы: ", reverse_arr[reverse_arr % 2 == 0], sep='\n')
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr = np.arange(0, 9).reshape(3, 3)
print(arr)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr_3d = np.random.rand(4, 3, 2)
print(arr_3d)
print("Min: ", arr_3d.min(), "Max: ", arr_3d.max())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
A = np.random.randint(0, 10, (6, 4))
B = np.random.randint(0, 10, (4, 3))
print("A: ", A, sep='\n')
print("B: ", B, sep='\n')
print("A @ B: ", A @ B, sep='\n')
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr = np.random.rand(7, 7).round(2)
print(arr)
mean = arr.mean()
std = arr.std()
print("Mean: ", mean, "Std: ", std)
arr_normalized = ((arr - mean) / std).round(2)
print(arr_normalized)
print("Mean: ", arr_normalized.mean(), "Std: ", arr_normalized.std())
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.hist(tips['total_bill'], bins=20)
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title('Total bill vs Tip')
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
sns.boxplot(x='day', y='total_bill', data=tips)                                                                                                                                       
plt.title('Взаимосвязь между total_bill и day')                                                                                                                                       
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
g = sns.FacetGrid(tips, col='time', height=5)                                                                                                                                         
g.map(plt.hist, 'tip', bins=20)                                                                                                                                                     
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
male_tips = tips[tips['sex'] == 'Male']
female_tips = tips[tips['sex'] == 'Female']

scatter1 = axes[0].scatter(male_tips['total_bill'], male_tips['tip'], c=male_tips['smoker'], cmap='coolwarm', alpha=0.6)
axes[0].set_title('Male')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')

scatter2 = axes[1].scatter(female_tips['total_bill'], female_tips['tip'], c=female_tips['smoker'], cmap='coolwarm', alpha=0.6)
axes[1].set_title('Female')
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Tip')

plt.colorbar(scatter2, ax=axes[1], label='Smoker (0=No, 1=Yes)')
plt.tight_layout()

plt.show()
```
