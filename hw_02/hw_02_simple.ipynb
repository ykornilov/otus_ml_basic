## Задание 1

Написать программу, которая получает на вход строку и возвращает словарь, где:
- ключи — символы из этой строки;
- значения — количество раз, сколько каждый символ встречается.

```python
user_input = input('Введите строку: ')

result = {}
for char in user_input:
    if char.isalnum():
        result[char] = result.get(char, 0) + 1
print(f'Словарь для строки "{user_input}": {result}')
```

## Задание 2

Дана строка текста (или введённая через консоль). Программа должна вернуть новую строку, в которой порядок слов будет обратным.
Пример:
"Python is really cool" → "cool really is Python".


```python
user_input = input('Введите строку: ')

words = user_input.split()
result = ' '.join(words[::-1])
print(f'"{user_input}" -> "{result}"')
```

## Задание 3

Написать программу, которая удаляет из списка все дубликаты, сохранив исходный порядок элементов.


```python
user_input = input("Введите список элементов через пробел: ")

words = user_input.split()
result = []
keys = set()
for word in words:
    if word not in keys:
        keys.add(word)
        result.append(word)
print(result)
```

## Задание 4

Даны три (или больше) списка с объектами. Программа должна создать новый список, содержащий все уникальные элементы — каждый объект встречается только один раз.


```python
lists = [
    [1, 2, 3, 4, 5],
    [2, 2, 3, 'ab', 'c'],
    ['a', 0, 1, 'c', 'b']
]

result_set = set()
for ls in lists:
    result_set = result_set.union(ls)
result = list(result_set)
print(result)
```

## Задание 5

Дана строка текста (или введённая через консоль). Программа должна вернуть словарь с четырьмя ключами:
- "гласные",
- "согласные",
- "цифры",
- "пунктуация".

Значения — количество символов каждого типа в строке.


```python
user_input = input('Введите строку: ').strip()

VOWELS = 'Гласные'
CONSONANTS = 'Согласные'
DIGITS = 'Цифры'
PUNCTUATIONS = 'Знаки пунктуации'

result = {
    VOWELS: 0,
    CONSONANTS: 0,
    DIGITS: 0,
    PUNCTUATIONS: 0,
}

vowel_set = {'а', 'е', 'ё', 'и', 'о', 'у', 'ы', 'э', 'ю', 'я'}
punctuation_set = {'.', ',', ';', ':', '!', '?', '-', '—', '«', '»', '(', ')', '[', ']', '{', '}'}

result[PUNCTUATIONS] += user_input.count('...')

for char in user_input.replace('...', ''):
    if char.isdecimal():
        result[DIGITS] += 1
    elif char.isalpha() and char.lower() in vowel_set:
        result[VOWELS] += 1
    elif char.isalpha():
        result[CONSONANTS] += 1
    elif char in punctuation_set:
        result [PUNCTUATIONS] += 1

for key, value in result.items():
    print(f'{key}: {value}')
```
